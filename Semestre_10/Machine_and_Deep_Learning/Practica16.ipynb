#### Practica 16 Implemente en Python Componentes Matemáticos del Deep Learning

#### Paso 1: Importar las librerias necesarias

```python
import numpy as np
```

#### Paso 2: Escalares - Tensores 0D
#### Un escalar es un solo numero. Es un tensor de 0 dimensiones (0D). En Python, un escalar se puede representar como un número simple.

```python
x = np.array(2)
x
```

#### Paso 3: Vectores - Tensores 1D
#### Un vector es una lista de números. Es un tensor de 1 dimensión (1D). En Python, un vector se puede representar como una lista o un array de una sola dimensión.

```python
x = np.array([12, 3, 6, 14])
x
```

#### Paso 4: Matrices - Tensores 2D
#### Una matriz es una tabla de números con filas y columnas. Es un tensor de 2 dimensiones (2D). En Python, una matriz se puede representar como una lista de listas o un array de dos dimensiones.

```python
x = np.array([[7, 80, 4, 38, 2],
              [4, 77, 1, 33, 3],
              [5, 78, 2, 34, 1]])
x
```

#### Paso 5: Tensores 3D y superiores
#### Un tensor de 3 dimensiones (3D) es una colección de matrices. En Python, un tensor de 3D se puede representar como una lista de listas de listas o un array de tres dimensiones. Los tensores de dimensiones superiores siguen el mismo patrón, con cada dimensión adicional representando una nueva "capa" de datos.

```python
x = np.array([[[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]],
              [[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]],
              [[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]]])
x
```

#### Paso 6: Tensores 4D y superiores
#### Un tensor de 4 dimensiones (4D) es una colección de tensores de 3D. En Python, un tensor de 4D se puede representar como una lista de listas de listas de listas o un array de cuatro dimensiones. Los tensores de dimensiones superiores siguen el mismo patrón, con cada dimensión adicional representando una nueva "capa" de datos.

```python
x = np.array([[[[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]],
              [[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]],
              [[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]]],
              [[[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]],
              [[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]],
              [[7, 80, 4, 38, 2], [4, 77, 1, 33, 3], [5, 78, 2, 34, 1]]]])
x
```
